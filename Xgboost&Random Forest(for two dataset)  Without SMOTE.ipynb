```python
import time
from sklearn.model_selection import train_test_split
import pandas as pd
import xgboost as xgb
from sklearn.metrics import (
accuracy_score, classification_report, confusion_matrix, precision_recall_fscore_support, precision_score, recall_score, f1_score
)
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming X_scaled and y_encoded are your scaled features and encoded labels

# Step 1: Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

# Display class distribution for Training and Test Sets
def show_class_distribution(y_train, y_test):
    train_counts = pd.Series(y_train).value_counts().sort_index()
    test_counts = pd.Series(y_test).value_counts().sort_index()

    print("Class Distribution in Training Set:")
    print(train_counts)
    print("\nClass Distribution in Test Set:")
    print(test_counts)

# Call the function to show class distribution
show_class_distribution(y_train, y_test)


# Function to evaluate the model (Accuracy, Classification Report, Confusion Matrix)
def evaluate_model(y_test, y_pred, model_name):
    # Accuracy
    accuracy = accuracy_score(y_test, y_pred)
    print(f"{model_name} Accuracy: {accuracy:.4f}")

    # Overall Precision, Recall, F1-Score
    precision = precision_score(y_test, y_pred, average='macro', zero_division=0)
    recall = recall_score(y_test, y_pred, average='macro', zero_division=0)
    f1 = f1_score(y_test, y_pred, average='macro', zero_division=0)
    print(f"{model_name} Precision: {precision:.4f}")
    print(f"{model_name} Recall: {recall:.4f}")
    print(f"{model_name} F1-Score: {f1:.4f}")

    # Classification Report
    class_report = classification_report(y_test, y_pred)
    print(f"\n{model_name} Classification Report:\n", class_report)
    
    # Confusion Matrix
    conf_matrix = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(10, 7))
    sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues")
    plt.title(f"{model_name} Confusion Matrix")
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.show()
    
    # Precision, Recall, F1 for each class
    precision, recall, f1, _ = precision_recall_fscore_support(y_test, y_pred, average=None)
    print(f"\n{model_name} Precision, Recall, F1 for each class:")
    for i, (p, r, f) in enumerate(zip(precision, recall, f1)):
        print(f"Class {i}: Precision={p:.4f}, Recall={r:.4f}, F1-Score={f:.4f}")

# Step 2: Define XGBoost parameters
params = {
    'objective': 'multi:softmax',  # For multi-class classification
    'num_class': len(set(y_train)),  # Number of unique classes
    'max_depth': 6,  # Maximum tree depth
    'learning_rate': 0.1,  # Learning rate
    'eta': 0.3,  # Step size shrinkage
    'eval_metric': 'mlogloss',  # Evaluation metric
    'seed': 42  # Random seed for reproducibility
}

##########################################
# Model: Train XGBoost WITHOUT Sampling
##########################################
print("Training XGBoost without Sampling...")

# Convert training and test data to DMatrix format
dtrain_no_sampling = xgb.DMatrix(X_train, label=y_train)
dtest = xgb.DMatrix(X_test, label=y_test)

# Train XGBoost without Sampling
xgb_model_no_sampling = xgb.train(params, dtrain_no_sampling, num_boost_round=100, evals=[(dtrain_no_sampling, 'train'), (dtest, 'test')],
                                  early_stopping_rounds=10, verbose_eval=True)

# Make predictions on the test set (without Sampling)
y_pred_no_sampling = xgb_model_no_sampling.predict(dtest).astype(int)

# Evaluate Model (without Sampling)
evaluate_model(y_test, y_pred_no_sampling, "XGBoost without Sampling")

```

```python
#### Install and import necessary libraries
!pip install scikit-learn
import time
import numpy as np
import pandas as pd
import psutil  # For memory monitoring
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix, 
    roc_curve, auc, precision_recall_curve, precision_score, 
    recall_score, f1_score
)
import matplotlib.pyplot as plt
import seaborn as sns
import xgboost as xgb
from xgboost import XGBClassifier

# Assume X_train, X_test, y_train, y_test are already defined

# Initialize and train Random Forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Train the XGBoost model using XGBClassifier
xgb_model_no_sampling = XGBClassifier(random_state=42)
xgb_model_no_sampling.fit(X_train, y_train)

# Step 1: Make predictions on the test set and calculate probabilities
y_pred_rf = rf_model.predict(X_test)
y_proba_rf = rf_model.predict_proba(X_test)
y_pred_xgb = xgb_model_no_sampling.predict(X_test)
y_proba_xgb = xgb_model_no_sampling.predict_proba(X_test)


# Calculate and display AUC for each model
fpr_rf, tpr_rf, _ = roc_curve(y_test, y_proba_rf[:, 1], pos_label=1)
auc_rf = auc(fpr_rf, tpr_rf)

fpr_xgb, tpr_xgb, _ = roc_curve(y_test, y_proba_xgb[:, 1], pos_label=1)
auc_xgb = auc(fpr_xgb, tpr_xgb)

print(f"Random Forest - AUC : {auc_rf:.4f}")
print(f"XGBoost - AUC : {auc_xgb:.4f}")

# Step 2: Confusion Matrix Visualization for Random Forest
conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)
plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix_rf, annot=True, fmt="d", cmap="Blues", xticklabels=np.unique(y_test), yticklabels=np.unique(y_test))
plt.title("Random Forest Without Sampling Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()

# Step 3: Precision-Recall Bar Chart (Per Class)
def plot_precision_recall_bar_chart(y_true, y_pred_rf, y_pred_xgb):
    precision_rf = precision_score(y_true, y_pred_rf, average=None, zero_division=0)
    recall_rf = recall_score(y_true, y_pred_rf, average=None, zero_division=0)
    precision_xgb = precision_score(y_true, y_pred_xgb, average=None, zero_division=0)
    recall_xgb = recall_score(y_true, y_pred_xgb, average=None, zero_division=0)
    
    classes = np.unique(y_true)
    bar_width = 0.35
    index = np.arange(len(classes))
    
    plt.figure(figsize=(15, 8))
    plt.bar(index, precision_rf, bar_width, label='Precision (RF)', alpha=0.6)
    plt.bar(index + bar_width, precision_xgb, bar_width, label='Precision (XGBoost)', alpha=0.6)
    plt.bar(index, recall_rf, bar_width, label='Recall (RF)', alpha=0.6, bottom=precision_rf)
    plt.bar(index + bar_width, recall_xgb, bar_width, label='Recall (XGBoost)', alpha=0.6, bottom=precision_xgb)
    
    plt.xlabel('Class')
    plt.ylabel('Score')
    plt.title('Precision and Recall Per Class')
    plt.xticks(index + bar_width / 2, classes)
    plt.legend()
    plt.show()

plot_precision_recall_bar_chart(y_test, y_pred_rf, y_pred_xgb)

# Step 4: Multiclass ROC Curve and AUC Scores for Each Model
def plot_multiclass_roc_curve(models, X_test, y_test):
    n_classes = len(np.unique(y_test))
    line_styles = ['-', '--', '-.', ':', (0, (3, 1, 1, 1)), (0, (5, 5))]

    # Determine the number of rows and columns for subplots
    n_cols = 3  # Adjust the number of columns as needed
    n_rows = (n_classes + n_cols - 1) // n_cols  # Calculate required rows

    fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, n_rows * 4))
    axes = axes.ravel()  # Flatten axes for easy iteration

    for i in range(n_classes):
        for model_name, model in models.items():
            y_proba = model.predict_proba(X_test)
            fpr, tpr, _ = roc_curve(y_test == i, y_proba[:, i])
            roc_auc = auc(fpr, tpr)

            axes[i].plot(
                fpr, tpr, 
                label=f'{model_name} (AUC = {roc_auc:.2f})',
                linestyle=line_styles[i % len(line_styles)],
                linewidth=2
            )
            axes[i].plot([0, 1], [0, 1], 'k--', lw=1)  # Diagonal line for reference

        axes[i].set_xlim([0.0, 1.0])
        axes[i].set_ylim([0.0, 1.05])
        axes[i].set_title(f'ROC Curve for Class {i}')
        axes[i].set_xlabel('False Positive Rate')
        axes[i].set_ylabel('True Positive Rate')
        axes[i].legend(loc="lower right")

    # Remove unused subplots
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.show()

# Call the function with the updated code
models = {
    "Random Forest": rf_model,
    "XGBoost": xgb_model_no_sampling
}
plot_multiclass_roc_curve(models, X_test, y_test)


# Step 5: Per-Class Precision, Recall, and F1 Score Line Plot
def plot_per_class_metrics(y_true, y_pred_rf, y_pred_xgb):
    precision_rf = precision_score(y_true, y_pred_rf, average=None, zero_division=0)
    recall_rf = recall_score(y_true, y_pred_rf, average=None, zero_division=0)
    f1_rf = f1_score(y_true, y_pred_rf, average=None, zero_division=0)
    
    precision_xgb = precision_score(y_true, y_pred_xgb, average=None, zero_division=0)
    recall_xgb = recall_score(y_true, y_pred_xgb, average=None, zero_division=0)
    f1_xgb = f1_score(y_true, y_pred_xgb, average=None, zero_division=0)
    
    classes = np.unique(y_true)
    
    plt.figure(figsize=(15, 8))
    plt.plot(classes, precision_rf, label='Precision (RF)', marker='o')
    plt.plot(classes, recall_rf, label='Recall (RF)', marker='o')
    plt.plot(classes, f1_rf, label='F1 Score (RF)', marker='o')
    
    plt.plot(classes, precision_xgb, label='Precision (XGBoost)', marker='x')
    plt.plot(classes, recall_xgb, label='Recall (XGBoost)', marker='x')
    plt.plot(classes, f1_xgb, label='F1 Score (XGBoost)', marker='x')
    
    plt.xlabel('Class')
    plt.ylabel('Score')
    plt.title('Per-Class Precision, Recall, and F1 Score')
    plt.legend()
    plt.grid(True)
    plt.show()

plot_per_class_metrics(y_test, y_pred_rf, y_pred_xgb)

# Print a detailed classification report for both models
print("\nRandom Forest Classification Report:\n", classification_report(y_test, y_pred_rf))
print("\nXGBoost Classification Report:\n", classification_report(y_test, y_pred_xgb))


# Calculate accuracy for both models
accuracy_rf = accuracy_score(y_test, y_pred_rf)
accuracy_xgb = accuracy_score(y_test, y_pred_xgb)

# Calculate precision, recall, and f1-score for both models using average='macro' for overall performance
precision_rf_macro = precision_score(y_test, y_pred_rf, average='macro')
recall_rf_macro = recall_score(y_test, y_pred_rf, average='macro')
f1_rf_macro = f1_score(y_test, y_pred_rf, average='macro')

precision_xgb_macro = precision_score(y_test, y_pred_xgb, average='macro')
recall_xgb_macro = recall_score(y_test, y_pred_xgb, average='macro')
f1_xgb_macro = f1_score(y_test, y_pred_xgb, average='macro')

# Print results
print(f"Random Forest Accuracy: {accuracy_rf:.4f}")
print(f"Random Forest Precision: {precision_rf_macro:.4f}")
print(f"Random Forest Recall: {recall_rf_macro:.4f}")
print(f"Random Forest F1 Score: {f1_rf_macro:.4f}")

print(f"\nXGBoost Accuracy: {accuracy_xgb:.4f}")
print(f"XGBoost Precision: {precision_xgb_macro:.4f}")
print(f"XGBoost Recall: {recall_xgb_macro:.4f}")
print(f"XGBoost F1 Score: {f1_xgb_macro:.4f}")
```
