```python
import numpy as np
import pandas as pd

# Automatically detect numerical columns
numerical_cols = combined_dataset.select_dtypes(include=['float64', 'int64']).columns

# Check for NaN values
print("NaN values in dataset:\n", combined_dataset[numerical_cols].isna().sum())

# Check for infinity values
print("Infinity values in dataset:\n", np.isinf(combined_dataset[numerical_cols]).sum())

# Fill NaN with the mean of each column
combined_dataset[numerical_cols] = combined_dataset[numerical_cols].fillna(combined_dataset[numerical_cols].mean())

# Replace infinity with a large finite number
combined_dataset[numerical_cols] = combined_dataset[numerical_cols].replace([np.inf, -np.inf], np.nan)
combined_dataset[numerical_cols] = combined_dataset[numerical_cols].fillna(combined_dataset[numerical_cols].mean())
```

```python
# Ensure that X_scaled (scaled features) and y (target) are prepared from previous steps
X = combined_dataset.drop(' Label', axis=1)  # Assuming 'Label' is the target column
y = combined_dataset[' Label']  # Target variable

# Encode the labels (target variable) using LabelEncoder
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# If needed, scale the features (assuming scaling is required)
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Confirm that all features are now numeric
print(combined_dataset.dtypes)

# Verify the scaling worked
print("First 5 rows of scaled features:\n", X_scaled[:5])
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.utils import resample
import numpy as np

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Start timing
start = time.time()
for ratio in smote_ratios:
    # Apply SMOTE with custom ratios
    smote = SMOTE(sampling_strategy={
        minority_classes.keys()[0]: int(ratio[0]),
        minority_classes.keys()[1]: int(ratio[1]),
        minority_classes.keys()[2]: int(ratio[2])
    }, k_neighbors=3, random_state=42)
    X_resampled, y_resampled = smote.fit_resample(X_train_B, y_train_B)
    
    # Resample to a smaller subset
    X_resampled_small, y_resampled_small = resample(X_resampled, y_resampled, n_samples=10000, random_state=42)
    # Encode target labels
    y_resampled_small_encoded = label_encoder.fit_transform(y_resampled_small)
    
    # Train XGBoost classifier
    xgb_classifier = XGBClassifier(eval_metric='logloss', random_state=42)
    xgb_classifier.fit(X_resampled_small, y_resampled_small_encoded)
    
    # Make predictions on validation set
    y_pred_xgb_encoded = xgb_classifier.predict(X_validation)
    y_pred_xgb = label_encoder.inverse_transform(y_pred_xgb_encoded)
    
    # Evaluate recall
    recall_xgb = recall_score(y_validation, y_pred_xgb, average=None)
    recall_scores_xgb.append(recall_xgb)
    rms_xgb = np.sqrt(np.mean(np.square(recall_xgb)))
    rms_errors_xgb.append(rms_xgb)

end = time.time()
print(end - start, ": finished XGBoost")
```

```python
# Determine the best SMOTE ratio for each classifier
best_ratio_rf = smote_ratios[np.argmin(rms_errors_rf)]
best_ratio_xgb = smote_ratios[np.argmin(rms_errors_xgb)]

print("Optimal SMOTE Ratio for Random Forest: \n", minority_classes.keys()[0], "= ", int(best_ratio_rf[0]), 
      "\n", minority_classes.keys()[1], "=", int(best_ratio_rf[1]), 
      "\n", minority_classes.keys()[2], "=", int(best_ratio_rf[2]))

print("Optimal SMOTE Ratio for XGBoost: \n", minority_classes.keys()[0], "= ", int(best_ratio_xgb[0]), 
      "\n", minority_classes.keys()[1], "=", int(best_ratio_xgb[1]), 
      "\n", minority_classes.keys()[2], "=", int(best_ratio_xgb[2]))
```

```python
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, 
    confusion_matrix, classification_report, roc_curve, auc
)
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Make predictions
y_pred_rf = final_rf_classifier.predict(X_test)
y_proba_rf = final_rf_classifier.predict_proba(X_test)
y_pred_xgb = final_xgb_classifier.predict(X_test)
y_proba_xgb = final_xgb_classifier.predict_proba(X_test)
# Encode y_test to integers (if y_test is strings)
y_test_encoded = label_encoder.transform(y_test)


# Display accuracy, precision, recall, and f1-score for each model
def display_metrics(model_name, y_true, y_pred):
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, average='macro', zero_division=0)
    recall = recall_score(y_true, y_pred, average='macro', zero_division=0)
    f1 = f1_score(y_true, y_pred, average='macro', zero_division=0)
    
    print(f"{model_name} - Accuracy: {accuracy:.4f}")
    print(f"{model_name} - Precision: {precision:.4f}")
    print(f"{model_name} - Recall: {recall:.4f}")
    print(f"{model_name} - F1-Score: {f1:.4f}\n")

# Evaluate metrics for RF, XGBoost
display_metrics("Random Forest", y_test, y_pred_rf)
display_metrics("XGBoost", y_test_encoded, y_pred_xgb)
#display_metrics("Random Forest", y_test, y_pred_rf)
#display_metrics("XGBoost", y_test, y_pred_xgb)

# AUC calculation for each model
y_test_bin = pd.get_dummies(y_test)  # Convert to binary for multiclass ROC AUC calculation
auc_rf = roc_auc_score(y_test_bin, y_proba_rf, multi_class='ovr', average='weighted')
auc_xgb = roc_auc_score(y_test_bin, y_proba_xgb, multi_class='ovr', average='weighted')
print(f"Random Forest - AUC: {auc_rf:.4f}")
print(f"XGBoost - AUC: {auc_xgb:.4f}")

# Confusion Matrix Visualization
def plot_confusion_matrix(model_name, y_true, y_pred):
    conf_matrix = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(10, 7))
    sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", 
                xticklabels=np.unique(y_true), yticklabels=np.unique(y_true))
    plt.title(f"{model_name} Confusion Matrix")
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.show()

plot_confusion_matrix("Random Forest", y_test, y_pred_rf)
plot_confusion_matrix("XGBoost", y_test_encoded, y_pred_xgb)

# Precision-Recall Bar Chart
def plot_precision_recall_bar_chart(y_true, y_pred_rf, y_pred_xgb):
    precision_rf = precision_score(y_true, y_pred_rf, average=None, zero_division=0)
    recall_rf = recall_score(y_true, y_pred_rf, average=None, zero_division=0)
    precision_xgb = precision_score(y_true, y_pred_xgb, average=None, zero_division=0)
    recall_xgb = recall_score(y_true, y_pred_xgb, average=None, zero_division=0)
    
    classes = np.unique(y_true)
    bar_width = 0.35
    index = np.arange(len(classes))
    
    plt.figure(figsize=(15, 8))
    plt.bar(index, precision_rf, bar_width, label='Precision (RF)', alpha=0.6)
    plt.bar(index + bar_width, precision_xgb, bar_width, label='Precision (XGBoost)', alpha=0.6)
    plt.bar(index, recall_rf, bar_width, label='Recall (RF)', alpha=0.6, bottom=precision_rf)
    plt.bar(index + bar_width, recall_xgb, bar_width, label='Recall (XGBoost)', alpha=0.6, bottom=precision_xgb)
    
    plt.xlabel('Class')
    plt.ylabel('Score')
    plt.title('Precision and Recall Per Class')
    plt.xticks(index + bar_width / 2, classes)
    plt.legend()
    plt.show()

y_pred_rf_encoded = label_encoder.transform(y_pred_rf)
plot_precision_recall_bar_chart(y_test_encoded, y_pred_rf_encoded, y_pred_xgb)
```

```python
# Multiclass ROC Curve
def plot_multiclass_roc_curve(models, X_test, y_test):
    n_classes = len(np.unique(y_test))
    line_styles = ['-', '--', '-.', ':', (0, (3, 1, 1, 1)), (0, (5, 5))]

    fig, axes = plt.subplots((n_classes + 2) // 3, 3, figsize=(18, 4 * ((n_classes + 2) // 3)))
    axes = axes.ravel()

    for i in range(n_classes):
        for model_name, model in models.items():
            y_proba = model.predict_proba(X_test)
            fpr, tpr, _ = roc_curve(y_test == i, y_proba[:, i])
            roc_auc = auc(fpr, tpr)
            axes[i].plot(fpr, tpr, label=f'{model_name} (AUC = {roc_auc:.4f})', linestyle=line_styles[i % len(line_styles)], linewidth=2)
            axes[i].plot([0, 1], [0, 1], 'k--', lw=1)

        axes[i].set_title(f'ROC Curve for Class {i}')
        axes[i].set_xlabel('False Positive Rate')
        axes[i].set_ylabel('True Positive Rate')
        axes[i].legend(loc="lower right")

    plt.tight_layout()
    plt.show()

models = {
    "Random Forest": final_rf_classifier,
    "XGBoost": final_xgb_classifier
}
plot_multiclass_roc_curve(models, X_test, y_test)
```

```python
# Per-Class Metrics Line Plot
def plot_per_class_metrics(y_true, y_pred_rf, y_pred_xgb):
    precision_rf = precision_score(y_true, y_pred_rf, average=None, zero_division=0)
    recall_rf = recall_score(y_true, y_pred_rf, average=None, zero_division=0)
    f1_rf = f1_score(y_true, y_pred_rf, average=None, zero_division=0)
    
    precision_xgb = precision_score(y_true, y_pred_xgb, average=None, zero_division=0)
    recall_xgb = recall_score(y_true, y_pred_xgb, average=None, zero_division=0)
    f1_xgb = f1_score(y_true, y_pred_xgb, average=None, zero_division=0)
    
    classes = np.unique(y_true)
    
    plt.figure(figsize=(15, 8))
    plt.plot(classes, precision_rf, label='Precision (RF)', marker='o')
    plt.plot(classes, recall_rf, label='Recall (RF)', marker='o')
    plt.plot(classes, f1_rf, label='F1 Score (RF)', marker='o')
    
    plt.plot(classes, precision_xgb, label='Precision (XGBoost)', marker='x')
    plt.plot(classes, recall_xgb, label='Recall (XGBoost)', marker='x')
    plt.plot(classes, f1_xgb, label='F1 Score (XGBoost)', marker='x')
    
    plt.xlabel('Class')
    plt.ylabel('Score')
    plt.title('Per-Class Precision, Recall, and F1 Score with SMOTE')
    plt.legend()
    plt.grid(True)
    plt.show()

y_pred_rf_encoded = label_encoder.transform(y_pred_rf)
plot_per_class_metrics(y_test_encoded, y_pred_rf_encoded, y_pred_xgb)

# Print classification reports
print("\nRandom Forest Classification Report:\n", classification_report(y_test, y_pred_rf))
print("\nXGBoost Classification Report:\n", classification_report(y_test_encoded, y_pred_xgb))
```
