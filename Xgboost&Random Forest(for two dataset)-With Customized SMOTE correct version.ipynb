```python
import time
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
import pandas as pd
import psutil  # For memory monitoring
import xgboost as xgb
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (accuracy_score, classification_report, confusion_matrix, roc_auc_score,
                             roc_curve, auc, precision_recall_curve, precision_recall_fscore_support, precision_score, recall_score, f1_score)
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.sparse
from collections import Counter

# Monitor memory usage
def check_memory():
    mem = psutil.virtual_memory()
    print(f"Memory usage: {mem.percent}% used out of {mem.total / (1024 ** 3):.2f} GB")

# Step 1: Split the data first, then apply SMOTE on the training set only
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)
# Print the exact number of samples in each set
print(f"Total number of samples in Training Set: {len(X_train)}")
print(f"Total number of samples in Test Set: {len(X_test)}")

# Step 2: Check the original class distribution
original_class_counts = Counter(y_train)
print(f"Original training class distribution: {original_class_counts}")

# Step 3: Apply SMOTE to the training set only with a valid multi-class sampling strategy
sampling_strategy = {cls: max(count, 5000) for cls, count in original_class_counts.items()}  # Adjust according to original counts
smote = SMOTE(sampling_strategy=sampling_strategy, random_state=42, k_neighbors=7)

# Apply SMOTE only to the training data
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Check the new class distribution after SMOTE
print(f"Resampled training class distribution: {Counter(y_train_resampled)}")

# Step 4: Convert the resampled data to sparse matrices
X_train_sparse = scipy.sparse.csr_matrix(X_train_resampled)
X_test_sparse = scipy.sparse.csr_matrix(X_test)

# Step 5: Convert Sparse Matrices to XGBoost DMatrix format
dtrain = xgb.DMatrix(X_train_sparse, label=y_train_resampled)
dtest = xgb.DMatrix(X_test_sparse, label=y_test)

# Step 6: Set up parameters for XGBoost
params = {
    'objective': 'multi:softprob',  # For multi-class classification
    'num_class': len(set(y_train)),  # Number of unique classes
    'max_depth': 6,  # Maximum tree depth
    'learning_rate': 0.1,  # Learning rate
    'eta': 0.3,  # Step size shrinkage
    'eval_metric': 'mlogloss',  # Evaluation metric
    'seed': 42  # Random seed for reproducibility
}

# Step 7: Train the XGBoost model with early stopping and store the evaluation results
evals = [(dtrain, 'train'), (dtest, 'test')]
evals_result = {}  # To store evaluation results
xgb_model = xgb.train(params, dtrain, num_boost_round=100, evals=evals, 
                       early_stopping_rounds=10, verbose_eval=True, evals_result=evals_result)

# Step 8: Make predictions on the test set
y_proba_xgb = xgb_model.predict(dtest)  # Probability predictions
y_pred_xgb = y_proba_xgb.argmax(axis=1)  # Convert probabilities to class predictions

# Step 9: Evaluate the XGBoost model
accuracy_xgb = accuracy_score(y_test, y_pred_xgb)
print(f"XGBoost Accuracy: {accuracy_xgb:.4f}")

# Step 10: Calculate and print overall Precision, Recall, and F1-Score
precision_xgb = precision_score(y_test, y_pred_xgb, average='macro', zero_division=0)
recall_xgb = recall_score(y_test, y_pred_xgb, average='macro', zero_division=0)
f1_xgb = f1_score(y_test, y_pred_xgb, average='macro', zero_division=0)

print(f"XGBoost Macro Precision: {precision_xgb:.4f}")
print(f"XGBoost Macro Recall: {recall_xgb:.4f}")
print(f"XGBoost Macro F1-Score: {f1_xgb:.4f}")

# Calculate AUC for XGBoost (multiclass)
roc_auc_xgb = roc_auc_score(y_test, y_proba_xgb, multi_class='ovr')
print(f"XGBoost AUC: {roc_auc_xgb:.4f}")

# Step 11: Classification report for XGBoost
class_report_xgb = classification_report(y_test, y_pred_xgb)
print("\nXGBoost Classification Report:\n", class_report_xgb)

# Step 12: Confusion Matrix for XGBoost
conf_matrix_xgb = confusion_matrix(y_test, y_pred_xgb)

# Visualization of Confusion Matrix for XGBoost
plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix_xgb, annot=True, fmt="d", cmap="Blues", xticklabels=set(y_test), yticklabels=set(y_test))
plt.title("XGBoost with Customized SMOTE Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()

# Step 13: Feature Importance for XGBoost
plt.figure(figsize=(10, 6))
xgb.plot_importance(xgb_model, max_num_features=10, importance_type='weight')
plt.title("XGBoost Feature Importance (Top 10)")
plt.show()

# Step 14: Learning Curve Visualization for XGBoost
# Retrieve the loss values from evals_result
train_loss = evals_result['train']['mlogloss']  # Extract train loss values
test_loss = evals_result['test']['mlogloss']    # Extract test loss values

# Plot learning curves for training and testing
plt.figure(figsize=(10, 6))
plt.plot(train_loss, label='Train Loss', color='blue')
plt.plot(test_loss, label='Test Loss', color='orange')
plt.xlabel('Number of Rounds')
plt.ylabel('Log Loss')
plt.title('Learning Curve for XGBoost')
plt.legend()
plt.grid(True)
plt.show()
```

```python
!pip install scikit-learn
import time
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
import pandas as pd
import psutil  # For memory monitoring
import xgboost as xgb
from xgboost import XGBClassifier

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (accuracy_score, classification_report, confusion_matrix, 
                             roc_curve, auc, precision_recall_curve, 
                             precision_score, recall_score, f1_score, roc_auc_score)
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.sparse
from collections import Counter

# Initialize and train Random Forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train_resampled, y_train_resampled)  # Train on resampled data

# Train the XGBoost model using XGBClassifier
xgb_model_with_sampling = XGBClassifier(random_state=42)
xgb_model_with_sampling.fit(X_train_resampled, y_train_resampled)


# Step 15: Make predictions on the test set with Random Forest
y_proba_rf = rf_model.predict_proba(X_test)
y_proba_xgb = xgb_model_with_sampling.predict_proba(X_test)
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print(f"Random Forest Accuracy: {accuracy_rf:.4f}")

# Calculate and print precision, recall, and F1-score for Random Forest
precision_rf = precision_score(y_test, y_pred_rf, average='macro', zero_division=0)
recall_rf = recall_score(y_test, y_pred_rf, average='macro', zero_division=0)
f1_rf = f1_score(y_test, y_pred_rf, average='macro', zero_division=0)
print(f"Random Forest Precision: {precision_rf:.4f}, Recall: {recall_rf:.4f}, F1-Score: {f1_rf:.4f}")

# Calculate AUC for Random Forest (multiclass)
roc_auc_rf = roc_auc_score(y_test, y_proba_rf, multi_class='ovr')
print(f"Random Forest AUC: {roc_auc_rf:.4f}")

# Step 16: Classification report for Random Forest
class_report_rf = classification_report(y_test, y_pred_rf)
print("\nRandom Forest Classification Report:\n", class_report_rf)

# Step 17: Confusion Matrix for Random Forest
conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)

# Visualization of Confusion Matrix for Random Forest
conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)
plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix_rf, annot=True, fmt="d", cmap="Blues", xticklabels=np.unique(y_test), yticklabels=np.unique(y_test))
plt.title("Random Forest with Customized SMOTE Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()

# Step 3: Precision-Recall Bar Chart (Per Class)
def plot_precision_recall_bar_chart(y_true, y_pred_rf, y_pred_xgb):
    precision_rf = precision_score(y_true, y_pred_rf, average=None, zero_division=0)
    recall_rf = recall_score(y_true, y_pred_rf, average=None, zero_division=0)
    precision_xgb = precision_score(y_true, y_pred_xgb, average=None, zero_division=0)
    recall_xgb = recall_score(y_true, y_pred_xgb, average=None, zero_division=0)
    
    classes = np.unique(y_true)
    bar_width = 0.35
    index = np.arange(len(classes))
    
    plt.figure(figsize=(15, 8))
    plt.bar(index, precision_rf, bar_width, label='Precision (RF)', alpha=0.6)
    plt.bar(index + bar_width, precision_xgb, bar_width, label='Precision (XGBoost)', alpha=0.6)
    plt.bar(index, recall_rf, bar_width, label='Recall (RF)', alpha=0.6, bottom=precision_rf)
    plt.bar(index + bar_width, recall_xgb, bar_width, label='Recall (XGBoost)', alpha=0.6, bottom=precision_xgb)
    
    plt.xlabel('Class')
    plt.ylabel('Score')
    plt.title('Precision and Recall Per Class with Customized SMOTE')
    plt.xticks(index + bar_width / 2, classes)
    plt.legend()
    plt.show()

plot_precision_recall_bar_chart(y_test, y_pred_rf, y_pred_xgb)

# Step 4: Multiclass ROC Curve and AUC Scores for Each Model
def plot_multiclass_roc_curve(models, X_test, y_test):
    n_classes = len(np.unique(y_test))
    line_styles = ['-', '--', '-.', ':', (0, (3, 1, 1, 1)), (0, (5, 5))]

    n_cols = 3  # Adjust the number of columns as needed
    n_rows = (n_classes + n_cols - 1) // n_cols  # Calculate required rows

    fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, n_rows * 4))
    axes = axes.ravel()  # Flatten axes for easy iteration

    for i in range(n_classes):
        for model_name, model in models.items():
            y_proba = model.predict_proba(X_test)
            fpr, tpr, _ = roc_curve(y_test == i, y_proba[:, i])
            roc_auc = auc(fpr, tpr)

            axes[i].plot(
                fpr, tpr, 
                label=f'{model_name} (AUC = {roc_auc:.2f})',
                linestyle=line_styles[i % len(line_styles)],
                linewidth=2
            )
            axes[i].plot([0, 1], [0, 1], 'k--', lw=1)  # Diagonal line for reference

        axes[i].set_xlim([0.0, 1.0])
        axes[i].set_ylim([0.0, 1.05])
        axes[i].set_title(f'ROC Curve for Class {i}')
        axes[i].set_xlabel('False Positive Rate')
        axes[i].set_ylabel('True Positive Rate')
        axes[i].legend(loc="lower right")

    # Remove unused subplots
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.show()

# Call the function with the updated code
models = {
    "Random Forest": rf_model,
    "XGBoost": xgb_model_with_sampling
}
plot_multiclass_roc_curve(models, X_test, y_test)

# Step 5: Get predicted probabilities for both models
y_proba_rf = rf_model.predict_proba(X_test)
y_proba_xgb = xgb_model_with_sampling.predict_proba(X_test)

# Step 6: Calculate AUC for each model
# For Random Forest (multiclass)
roc_auc_rf = roc_auc_score(y_test, y_proba_rf, multi_class='ovr')
print(f"Random Forest AUC: {roc_auc_rf:.4f}")

# For XGBoost (multiclass)
roc_auc_xgb = roc_auc_score(y_test, y_proba_xgb, multi_class='ovr')
print(f"XGBoost AUC: {roc_auc_xgb:.4f}")


# Step 6: Per-Class Precision, Recall, and F1 Score Line Plot
def plot_per_class_metrics(y_true, y_pred_rf, y_pred_xgb):
    precision_rf = precision_score(y_true, y_pred_rf, average=None, zero_division=0)
    recall_rf = recall_score(y_true, y_pred_rf, average=None, zero_division=0)
    f1_rf = f1_score(y_true, y_pred_rf, average=None, zero_division=0)
    
    precision_xgb = precision_score(y_true, y_pred_xgb, average=None, zero_division=0)
    recall_xgb = recall_score(y_true, y_pred_xgb, average=None, zero_division=0)
    f1_xgb = f1_score(y_true, y_pred_xgb, average=None, zero_division=0)
    
    classes = np.unique(y_true)
    
    plt.figure(figsize=(15, 8))
    plt.plot(classes, precision_rf, label='Precision (RF)', marker='o')
    plt.plot(classes, recall_rf, label='Recall (RF)', marker='o')
    plt.plot(classes, f1_rf, label='F1 Score (RF)', marker='o')
    
    plt.plot(classes, precision_xgb, label='Precision (XGBoost)', marker='x')
    plt.plot(classes, recall_xgb, label='Recall (XGBoost)', marker='x')
    plt.plot(classes, f1_xgb, label='F1 Score (XGBoost)', marker='x')
    
    plt.xlabel('Class')
    plt.ylabel('Score')
    plt.title('Per-Class Precision, Recall, and F1 Score with Customized SMOTE')
    plt.legend()
    plt.grid(True)
    plt.show()

plot_per_class_metrics(y_test, y_pred_rf, y_pred_xgb)

# Print a detailed classification report for both models
print("\nRandom Forest Classification Report:\n", classification_report(y_test, y_pred_rf))
print("\nXGBoost Classification Report:\n", classification_report(y_test, y_pred_xgb))
```
