```python
import pandas as pd
import os
dataset_paths = [
    'C:/Jupyter notebook/dataset_1.csv',
    'C:/Jupyter notebook/dataset_3.csv',
]

# Check if files exist
for path in dataset_paths:
    if not os.path.exists(path):
        print(f"File not found: {path}")

# If all files are confirmed to exist, read them
dataframes = [pd.read_csv(path) for path in dataset_paths if os.path.exists(path)]

# Combine all datasets into one dataframe
combined_dataset = pd.concat(dataframes, ignore_index=True)
```

```python
# Handle missing values (option to impute or drop)
# Dropping rows with missing values
combined_dataset = combined_dataset.dropna()

# Remove duplicate records
combined_dataset = combined_dataset.drop_duplicates()

# Checking if duplicates were removed
print("Data shape after removing duplicates:", combined_dataset.shape)
```

```python
import numpy as np
import pandas as pd

# Automatically detect numerical columns
numerical_cols = combined_dataset.select_dtypes(include=['float64', 'int64']).columns

# Check for NaN values
print("NaN values in dataset:\n", combined_dataset[numerical_cols].isna().sum())

# Check for infinity values
print("Infinity values in dataset:\n", np.isinf(combined_dataset[numerical_cols]).sum())

# Fill NaN with the mean of each column
combined_dataset[numerical_cols] = combined_dataset[numerical_cols].fillna(combined_dataset[numerical_cols].mean())

# Replace infinity with a large finite number
combined_dataset[numerical_cols] = combined_dataset[numerical_cols].replace([np.inf, -np.inf], np.nan)
combined_dataset[numerical_cols] = combined_dataset[numerical_cols].fillna(combined_dataset[numerical_cols].mean())
```

```python
# Ensure that X_scaled (scaled features) and y (target) are prepared from previous steps
X = combined_dataset.drop(' Label', axis=1)  # Assuming 'Label' is the target column
y = combined_dataset[' Label']  # Target variable

# Encode the labels (target variable) using LabelEncoder
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# If needed, scale the features (assuming scaling is required)
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Confirm that all features are now numeric
print(combined_dataset.dtypes)

# Verify the scaling worked
print("First 5 rows of scaled features:\n", X_scaled[:5])
```

```python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.model_selection import train_test_split, GridSearchCV
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import SMOTE
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix, 
    precision_score, recall_score, f1_score, roc_curve, auc, roc_auc_score
)
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier

# Split the original data into training and test sets first
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_encoded, test_size=0.2, random_state=42)

# Function to find the optimal downsampling rate on the training set
def find_best_downsampling(X_train, y_train, sampling_rates=[0.5, 0.4, 0.3, 0.2, 0.1]):
    best_score = 0
    best_rate = None
    best_X_resampled, best_y_resampled = None, None

    for rate in sampling_rates:
        under_sampler = RandomUnderSampler(sampling_strategy={0: int(len(y_train[y_train == 0]) * rate)}, random_state=42)
        X_under, y_under = under_sampler.fit_resample(X_train, y_train)


        # Show class distribution after downsampling
        downsampled_distribution = pd.Series(y_under).value_counts()
        print(f"Class distribution after downsampling with rate {rate}:\n{downsampled_distribution}\n")
        
        smote = SMOTE(random_state=42)
        X_resampled, y_resampled = smote.fit_resample(X_under, y_under)


        # Show class distribution after SMOTE
        smote_distribution = pd.Series(y_resampled).value_counts()
        print(f"Class distribution after applying SMOTE with rate {rate}:\n{smote_distribution}\n")
        
        
        # Split resampled training data into a temporary train-validation set for evaluation
        X_train_temp, X_val, y_train_temp, y_val = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)
        
        model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42)
        model.fit(X_train_temp, y_train_temp)
        y_pred = model.predict(X_val)
        accuracy = accuracy_score(y_val, y_pred)
        
        if accuracy > best_score:
            best_score = accuracy
            best_rate = rate
            best_X_resampled, best_y_resampled = X_resampled, y_resampled

    print(f"Best downsampling rate: {best_rate} with accuracy: {best_score:.4f}")
    return best_X_resampled, best_y_resampled

# Apply downsampling and SMOTE only on the training set to find the best downsampling rate
X_resampled, y_resampled = find_best_downsampling(X_train, y_train)

# Check the new class distribution in the resampled training set
new_class_distribution = pd.Series(y_resampled).value_counts()
print("New Class Distribution after Downsampling and SMOTE:\n", new_class_distribution)


# Check class distribution in the training set
train_class_distribution = pd.Series(y_train).value_counts()
print("Class Distribution in Training Set:\n", train_class_distribution)

# Check class distribution in the testing set
test_class_distribution = pd.Series(y_test).value_counts()
print("Class Distribution in Testing Set:\n", test_class_distribution)

# Train models on the resampled training set
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_resampled, y_resampled)

xgb_model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42)
xgb_model.fit(X_resampled, y_resampled)


# Make predictions
y_pred_rf = rf_model.predict(X_test)
y_proba_rf = rf_model.predict_proba(X_test)
y_pred_xgb = xgb_model.predict(X_test)
y_proba_xgb = xgb_model.predict_proba(X_test)

# Display accuracy, precision, recall, and f1-score for each model
def display_metrics(model_name, y_true, y_pred):
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, average='macro', zero_division=0)
    recall = recall_score(y_true, y_pred, average='macro', zero_division=0)
    f1 = f1_score(y_true, y_pred, average='macro', zero_division=0)
    
    print(f"{model_name} - Accuracy: {accuracy:.4f}")
    print(f"{model_name} - Precision: {precision:.4f}")
    print(f"{model_name} - Recall: {recall:.4f}")
    print(f"{model_name} - F1-Score: {f1:.4f}\n")

display_metrics("Random Forest", y_test, y_pred_rf)
display_metrics("XGBoost", y_test, y_pred_xgb)

# AUC calculation for each model
y_test_bin = pd.get_dummies(y_test)  # Convert to binary for multiclass ROC AUC calculation
auc_rf = roc_auc_score(y_test_bin, y_proba_rf, multi_class='ovr')
auc_xgb = roc_auc_score(y_test_bin, y_proba_xgb, multi_class='ovr')
print(f"Random Forest - AUC: {auc_rf:.4f}")
print(f"XGBoost - AUC: {auc_xgb:.4f}")

# Step 1: Confusion Matrix Visualization
conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)
plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix_rf, annot=True, fmt="d", cmap="Blues", xticklabels=np.unique(y_test), yticklabels=np.unique(y_test))
plt.title("Random Forest with Downsampling and SMOTE Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()

# Confusion Matrix for XGBoost
conf_matrix_xgb = confusion_matrix(y_test, y_pred_xgb)
plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix_xgb, annot=True, fmt="d", cmap="Blues", xticklabels=np.unique(y_test), yticklabels=np.unique(y_test))
plt.title("XGBoost with Downsampling and SMOTE Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()

# Step 2: Precision-Recall Bar Chart
def plot_precision_recall_bar_chart(y_true, y_pred_rf, y_pred_xgb):
    precision_rf = precision_score(y_true, y_pred_rf, average=None, zero_division=0)
    recall_rf = recall_score(y_true, y_pred_rf, average=None, zero_division=0)
    precision_xgb = precision_score(y_true, y_pred_xgb, average=None, zero_division=0)
    recall_xgb = recall_score(y_true, y_pred_xgb, average=None, zero_division=0)
    
    classes = np.unique(y_true)
    bar_width = 0.35
    index = np.arange(len(classes))
    
    plt.figure(figsize=(15, 8))
    plt.bar(index, precision_rf, bar_width, label='Precision (RF)', alpha=0.6)
    plt.bar(index + bar_width, precision_xgb, bar_width, label='Precision (XGBoost)', alpha=0.6)
    plt.bar(index, recall_rf, bar_width, label='Recall (RF)', alpha=0.6, bottom=precision_rf)
    plt.bar(index + bar_width, recall_xgb, bar_width, label='Recall (XGBoost)', alpha=0.6, bottom=precision_xgb)
    
    plt.xlabel('Class')
    plt.ylabel('Score')
    plt.title('Precision and Recall Per Class')
    plt.xticks(index + bar_width / 2, classes)
    plt.legend()
    plt.show()

plot_precision_recall_bar_chart(y_test, y_pred_rf, y_pred_xgb)

# Step 3: Multiclass ROC Curve
def plot_multiclass_roc_curve(models, X_test, y_test):
    n_classes = len(np.unique(y_test))
    line_styles = ['-', '--', '-.', ':', (0, (3, 1, 1, 1)), (0, (5, 5))]

    fig, axes = plt.subplots((n_classes + 2) // 3, 3, figsize=(18, 4 * ((n_classes + 2) // 3)))
    axes = axes.ravel()

    for i in range(n_classes):
        for model_name, model in models.items():
            y_proba = model.predict_proba(X_test)
            fpr, tpr, _ = roc_curve(y_test == i, y_proba[:, i])
            roc_auc = auc(fpr, tpr)
            axes[i].plot(fpr, tpr, label=f'{model_name} (AUC = {roc_auc:.2f})', linestyle=line_styles[i % len(line_styles)], linewidth=2)
            axes[i].plot([0, 1], [0, 1], 'k--', lw=1)

        axes[i].set_title(f'ROC Curve for Class {i}')
        axes[i].set_xlabel('False Positive Rate')
        axes[i].set_ylabel('True Positive Rate')
        axes[i].legend(loc="lower right")

    plt.tight_layout()
    plt.show()

models = {
    "Random Forest": rf_model,
    "XGBoost": xgb_model
}
plot_multiclass_roc_curve(models, X_test, y_test)

# Step 4: Per-Class Metrics Line Plot
def plot_per_class_metrics(y_true, y_pred_rf, y_pred_xgb):
    precision_rf = precision_score(y_true, y_pred_rf, average=None, zero_division=0)
    recall_rf = recall_score(y_true, y_pred_rf, average=None, zero_division=0)
    f1_rf = f1_score(y_true, y_pred_rf, average=None, zero_division=0)
    
    precision_xgb = precision_score(y_true, y_pred_xgb, average=None, zero_division=0)
    recall_xgb = recall_score(y_true, y_pred_xgb, average=None, zero_division=0)
    f1_xgb = f1_score(y_true, y_pred_xgb, average=None, zero_division=0)
    
    classes = np.unique(y_true)
    
    plt.figure(figsize=(15, 8))
    plt.plot(classes, precision_rf, label='Precision (RF)', marker='o')
    plt.plot(classes, recall_rf, label='Recall (RF)', marker='o')
    plt.plot(classes, f1_rf, label='F1 Score (RF)', marker='o')
    
    plt.plot(classes, precision_xgb, label='Precision (XGBoost)', marker='x')
    plt.plot(classes, recall_xgb, label='Recall (XGBoost)', marker='x')
    plt.plot(classes, f1_xgb, label='F1 Score (XGBoost)', marker='x')
    
    plt.xlabel('Class')
    plt.ylabel('Score')
    plt.title('Per-Class Precision, Recall, and F1 Score with Downsampling and SMOTE')
    plt.legend()
    plt.grid(True)
    plt.show()

plot_per_class_metrics(y_test, y_pred_rf, y_pred_xgb)

# Print classification reports
print("\nRandom Forest Classification Report:\n", classification_report(y_test, y_pred_rf))
print("\nXGBoost Classification Report:\n", classification_report(y_test, y_pred_xgb))
```
