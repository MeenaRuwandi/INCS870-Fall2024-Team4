```python
import pandas as pd
import os
dataset_paths = [
    'C:/Jupyter notebook/dataset_1.csv',
    'C:/Jupyter notebook/dataset_2.csv',
    'C:/Jupyter notebook/dataset_3.csv',
    'C:/Jupyter notebook/dataset_4.csv',
    'C:/Jupyter notebook/dataset_5.csv',
    'C:/Jupyter notebook/dataset_6.csv',
    'C:/Jupyter notebook/dataset_7.csv',
    'C:/Jupyter notebook/dataset_8.csv',
]

# Check if files exist
for path in dataset_paths:
    if not os.path.exists(path):
        print(f"File not found: {path}")

# If all files are confirmed to exist, read them
dataframes = [pd.read_csv(path) for path in dataset_paths if os.path.exists(path)]

# Combine all datasets into one dataframe
combined_dataset = pd.concat(dataframes, ignore_index=True)
```

```python
# Check the distribution of the target class (assuming 'Label' is the target column)
print(combined_dataset[' Label'].value_counts())

# Identifying the minority classes
minority_classes = combined_dataset[' Label'].value_counts()[combined_dataset[' Label'].value_counts() < 1000]
print("Minority attack classes:\n", minority_classes)
```

```python
# Handle missing values (option to impute or drop)
# Dropping rows with missing values
combined_dataset = combined_dataset.dropna()

# Remove duplicate records
combined_dataset = combined_dataset.drop_duplicates()

# Checking if duplicates were removed
print("Data shape after removing duplicates:", combined_dataset.shape)
```

```python
import numpy as np
import pandas as pd

# Automatically detect numerical columns
numerical_cols = combined_dataset.select_dtypes(include=['float64', 'int64']).columns

# Check for NaN values
print("NaN values in dataset:\n", combined_dataset[numerical_cols].isna().sum())

# Check for infinity values
print("Infinity values in dataset:\n", np.isinf(combined_dataset[numerical_cols]).sum())

# Fill NaN with the mean of each column
combined_dataset[numerical_cols] = combined_dataset[numerical_cols].fillna(combined_dataset[numerical_cols].mean())

# Replace infinity with a large finite number
combined_dataset[numerical_cols] = combined_dataset[numerical_cols].replace([np.inf, -np.inf], np.nan)
combined_dataset[numerical_cols] = combined_dataset[numerical_cols].fillna(combined_dataset[numerical_cols].mean())

```

```python
# Ensure that X_scaled (scaled features) and y (target) are prepared from previous steps
X = combined_dataset.drop(' Label', axis=1)  # Assuming 'Label' is the target column
y = combined_dataset[' Label']  # Target variable

# Encode the labels (target variable) using LabelEncoder
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# If needed, scale the features (assuming scaling is required)
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Confirm that all features are now numeric
print(combined_dataset.dtypes)

# Verify the scaling worked
print("First 5 rows of scaled features:\n", X_scaled[:5])
```

```python
import xgboost as xgb
from xgboost import XGBClassifier
import matplotlib.pyplot as plt
import seaborn as sns

# Split the dataset for initial XGBoost training
from sklearn.model_selection import train_test_split
X_train_temp, X_test_temp, y_train_temp, y_test_temp = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Train XGBoost model to get feature importance
xgb_model_temp = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
xgb_model_temp.fit(X_train_temp, y_train_temp)

# Get feature importance and select the top 20
importances = xgb_model_temp.feature_importances_
feature_importances = pd.DataFrame({'feature': X.columns, 'importance': importances}).sort_values(by='importance', ascending=False)
top_20_features = feature_importances.head(20)['feature'].tolist()

# Reduce the dataset to the top 20 features
X_reduced = X[top_20_features]

# Visualize the top 20 features
plt.figure(figsize=(10, 6))
sns.barplot(x='importance', y='feature', data=feature_importances.head(20), palette='viridis')
plt.title('Top 20 Most Important Features')
plt.xlabel('Importance Score')
plt.ylabel('Features')
plt.show()
```

```python
import time
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
import pandas as pd
import psutil  # For memory monitoring
import xgboost as xgb
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (accuracy_score, classification_report, confusion_matrix, 
                             roc_curve, auc, precision_recall_curve)
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.sparse
from collections import Counter

# Monitor memory usage
def check_memory():
    mem = psutil.virtual_memory()
    print(f"Memory usage: {mem.percent}% used out of {mem.total / (1024 ** 3):.2f} GB")

# Step 1: Split the data first, then apply SMOTE on the training set only
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

# Step 2: Check the original class distribution
original_class_counts = Counter(y_train)
print(f"Original training class distribution: {original_class_counts}")

# Step 3: Apply SMOTE to the training set only with a valid multi-class sampling strategy
sampling_strategy = {cls: max(count, 5000) for cls, count in original_class_counts.items()}  # Adjust according to original counts
smote = SMOTE(sampling_strategy=sampling_strategy, random_state=42, k_neighbors=7)

# Apply SMOTE only to the training data
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Check the new class distribution after SMOTE
print(f"Resampled training class distribution: {Counter(y_train_resampled)}")

# Step 4: Convert the resampled data to sparse matrices
X_train_sparse = scipy.sparse.csr_matrix(X_train_resampled)
X_test_sparse = scipy.sparse.csr_matrix(X_test)

# Step 5: Convert Sparse Matrices to XGBoost DMatrix format
dtrain = xgb.DMatrix(X_train_sparse, label=y_train_resampled)
dtest = xgb.DMatrix(X_test_sparse, label=y_test)

# Step 6: Set up parameters for XGBoost
params = {
    'objective': 'multi:softmax',  # For multi-class classification
    'num_class': len(set(y_train)),  # Number of unique classes
    'max_depth': 6,  # Maximum tree depth
    'learning_rate': 0.1,  # Learning rate
    'eta': 0.3,  # Step size shrinkage
    'eval_metric': 'mlogloss',  # Evaluation metric
    'seed': 42  # Random seed for reproducibility
}

# Step 7: Train the XGBoost model with early stopping and store the evaluation results
evals = [(dtrain, 'train'), (dtest, 'test')]
evals_result = {}  # To store evaluation results
xgb_model = xgb.train(params, dtrain, num_boost_round=100, evals=evals, 
                       early_stopping_rounds=10, verbose_eval=True, evals_result=evals_result)

# Step 8: Make predictions on the test set
y_pred_xgb = xgb_model.predict(dtest)
y_pred_xgb = y_pred_xgb.astype(int)  # Convert predictions to integer

# Step 9: Evaluate the XGBoost model
accuracy_xgb = accuracy_score(y_test, y_pred_xgb)
print(f"XGBoost Accuracy: {accuracy_xgb:.4f}")

# Step 10: Classification report for XGBoost
class_report_xgb = classification_report(y_test, y_pred_xgb)
print("\nXGBoost Classification Report:\n", class_report_xgb)

# Step 11: Confusion Matrix for XGBoost
conf_matrix_xgb = confusion_matrix(y_test, y_pred_xgb)

# Visualization of Confusion Matrix for XGBoost
plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix_xgb, annot=True, fmt="d", cmap="Blues", xticklabels=set(y_test), yticklabels=set(y_test))
plt.title("XGBoost Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()

# Step 12: Feature Importance for XGBoost
plt.figure(figsize=(10, 6))
xgb.plot_importance(xgb_model, max_num_features=10, importance_type='weight')
plt.title("XGBoost Feature Importance (Top 10)")
plt.show()

# Step 13: Learning Curve Visualization for XGBoost
# Retrieve the loss values from evals_result
train_loss = evals_result['train']['mlogloss']  # Extract train loss values
test_loss = evals_result['test']['mlogloss']    # Extract test loss values

# Plot learning curves for training and testing
plt.figure(figsize=(10, 6))
plt.plot(train_loss, label='Train Loss', color='blue')
plt.plot(test_loss, label='Test Loss', color='orange')
plt.xlabel('Number of Rounds')
plt.ylabel('Log Loss')
plt.title('Learning Curve for XGBoost')
plt.legend()
plt.grid(True)
plt.show()

```
